## Alur Kerja Decoding (Biner ke String)

Kalian harus membaca gambar satu per satu sesuai urutan nomor filenya, lalu terjemahkan hasil prediksi model CNN kalian menjadi teks dengan aturan berikut:

A. Translasi Biner

- Prediksi GENAP $\rightarrow$ diterjemahkan menjadi bit 0.
- Prediksi GANJIL $\rightarrow$ diterjemahkan menjadi bit 1.
- Prediksi ANGKA 0 $\rightarrow$ dianggap sebagai Spasi (Pemisah Kata).

B. Konversi ASCII 8-bitSetiap 8 bit (8 gambar ganjil/genap) yang kalian kumpulkan akan membentuk 1 karakter. Jika kalian menemukan gambar angka 0, abaikan dan mulai kumpulkan 8 bit berikutnya untuk karakter baru.

Contoh:

- 8 gambar pertama menghasilkan biner 01000001 $\rightarrow$ Karakter 'A'

- Gambar ke-9 adalah angka 0 $\rightarrow$ Spasi.

- 8 gambar berikutnya menghasilkan biner 01000010 $\rightarrow$ Karakter 'B'.

Hasil Akhir: "A B"

Jadi kalian hasil prediksi model CNN kalian, kalian bikin rule sesuai dengan diatas. itu kan hasilnya biner, kalian terjemahkan. terus kalian lakukan dibawah ini:

```python
# contoh cara mengubah list biner menjadi karakter di Python
biner_contoh = "01000001"
karakter = chr(int(biner_contoh, 2))
print(karakter)
```
